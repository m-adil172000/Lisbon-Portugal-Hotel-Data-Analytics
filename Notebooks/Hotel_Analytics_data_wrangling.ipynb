
**About Dataset:**

This data article describes a hotel customer dataset with 31 variables describing a total of 83,590 instances (customers). It comprehends three full years of customer behavioral data. In addition to personal and behavioral information, the dataset also contains demographic and geographical information.


**Data Source:**

*Antonio, Nuno; de Almeida, Ana; Nunes, Luis (2020), “Lisbon, Portugal, hotel’s customer dataset with three years of personal, behavioral, demographic, and geographic information”, Mendeley Data, V1, doi: 10.17632/j83f5fsh6c.1*



A hotel's customers personal, behavioral, demographic, and geographic dataset from Lisbon, Portugal (2015–2018)



The unit of analysis of the dataset is a customer. A full three-year period of data was collected (2015 to 2018). All personal related data were transformed or anonymized to guaranty privacy and prevent the hotel or guests’ identification. Time-related variables were accounted for based on the last day of the extraction period. The last day of the extraction period is December 31, 2018.

**DaysSinceCreation**

- Since the last day of the extraction period is December 31, 2018. And we the values of days since customer record was created. We can calculate the date when the record was created.

```python
# DaysSinceCreation - Days since customer record was created.

# The last day of the extraction period is December 31, 2018.

data.DaysSinceCreation.value_counts()
```

```python
# Define the reference date
reference_date = pd.to_datetime("2018-12-31")

# Subtract days from reference date
creation_dates = reference_date - pd.to_timedelta(data["DaysSinceCreation"], unit="D")

creation_dates
```

- The values in the `DaysSinceCreation` column range from 0 to 1095.
- We create a new column CreationDate by subtracting the DaysSinceCreation from the reference date (e.g., December 31, 2018). This represents the date when the customer's record was first created.
- Since CreationDate corresponds to the moment the customer first entered the system (based on the context of the dataset), it is reasonable to interpret it as the first booking date for that customer.
- Therefore, CreationDate can be used to analyze first-time booking patterns, such as:
  - Trends in new customer acquisition over time,
  - Seasonality in first-time bookings,
  - Cohort analysis based on signup dates,
  - Time since first booking and its relationship to customer loyalty or revenue.



**AverageLeadTime**

**FINDINGS:**


* The `ID` column contains unique values and matches the number of rows — confirms one row per customer.
* In the `Age` column, the Missing values are **MCAR (Missing Completely At Random)**:

  * No significant change in `Age` distribution after dropping nulls.
  * Categorical distributions remain stable.
  * **Safe to drop rows with missing values**.
* We need to removed the rows where:

  * `Age` is negative or greater than 100.
  * `AverageLeadTime` is negative.
* Create `CreationDate` from `DaysSinceCreation` and reference date (Dec 31, 2018):

  * Interpreted as the **first booking date** for each customer.
* Create `AvgArrivalDate` by adding `AverageLeadTime` to `CreationDate`:

  * Represents the **typical arrival date** for that customer.
* Create `TotalRevenue` as sum of `LodgingRevenue` and `OtherRevenue`.
* Create `TotalBookings` by summing `BookingsCanceled`, `BookingsNoShowed`, and `BookingsCheckedIn`.
* Validate stay logic:

  * Ensure that `DaysSinceLastStay` ≤ `DaysSinceFirstStay`.
  * If `DaysSinceFirstStay = -1`, then ensure that `BookingsCheckedIn = 0`. If all bookings are 0, further analysis is required (possible invalid records).
* Create `FirstArrivalDate` and `LastArrivalDate` as an approximate of the first arrival and last arrival date of the customer using `DaysSinceFirstStay` and `DaysSinceLastStay` columns.
* Create `SpecialRequests` by adding all the `SR` columns.
* Merge `Nationality` column with external country code dataset to get full country names.
* Group **least frequent countries** under a new category `'Others'`.
* `NameHash`and `DocIDHash` columns are not relevant for our analysis.



### **Data Cleaning and Transformation:**

```python
def clean_data(df):

  return (
      df
      .dropna()
      .query("Age>=0 and Age<=100")
      .query("AverageLeadTime >= 0")
      .assign(
          CreationDate = lambda df_: pd.to_datetime("2018-12-31") - pd.to_timedelta(df_.DaysSinceCreation,unit="D"),
          AvgArrivalDate = lambda df_: df_.CreationDate + pd.to_timedelta(df_.AverageLeadTime, unit="D"),
          TotalRevenue = lambda df_: df_.LodgingRevenue + df_.OtherRevenue,
          TotalBookings = lambda df_: df_.BookingsCanceled + df_.BookingsNoShowed + df_.BookingsCheckedIn,
          SpecialRequests = lambda df_: df_[['SRHighFloor', 'SRLowFloor', 'SRAccessibleRoom', 'SRMediumFloor', 'SRBathtub', 'SRShower', 'SRCrib', 'SRKingSizeBed', 'SRTwinBed', 'SRNearElevator', 'SRAwayFromElevator', 'SRNoAlcoholInMiniBar', 'SRQuietRoom']].sum(axis=1),
          FirstArrivalDate = lambda df_: (pd.to_datetime("2018-12-31") - pd.to_timedelta(df_.DaysSinceFirstStay.mask(df_.DaysSinceFirstStay == -1), unit="D")).dt.date,
          LastArrivalDate = lambda df_: (pd.to_datetime("2018-12-31") - pd.to_timedelta(df_.DaysSinceLastStay.mask(df_.DaysSinceLastStay == -1), unit="D")).dt.date,

      )
      .fillna(-1)
      .merge(codes,how='left',left_on='Nationality',right_on='Alpha-3 code')
      .drop(columns=['Nationality','Alpha-3 code','NameHash','DocIDHash'])
  )
```

```python
data_cleaned = clean_data(data)

data_cleaned
```
